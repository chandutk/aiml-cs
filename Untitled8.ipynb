one sample t test

```python
import numpy as np
from scipy import stats
d=pd.read_csv('/content/fruits.zip')
height=np.array([7.3,8.6,7.2,4.6,4.7,8.4,8,7.4,6.2,6])
t_stat,p_value= stats.ttest_1samp(height,popmean=7.2)
print( f"one sample t-test:{t_stat},p_value:{p_value}")
```

chi-square

```python
import pandas as pd
from scipy.stats import chi2_contingency
data = pd.read_csv('/content/fruits.zip')
contingency_table = pd.crosstab(data['fruit_subtype'], data['fruit_name'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f"Chi-square Test of Independence: chi2={chi2},p={p},dof={dof},expected={ex}")
```

anova (analysis of veriance)

```python
import pandas as pd
from scipy.stats import f_oneway
d=pd.read_csv('/content/fruits.zip')
groups = [d['width'], d['height'], ]
f_stat, p_value = f_oneway(*groups)
print(f"F-valuei:{f_stat},p-value:{p_value}")
```

wilcoxon

```python
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon
d=pd.read_csv('/content/fruits.zip')
data={
    'before':[7.3,8.6,7.2,4.6,4.7,8.4,8,7.4,6.2,6],
    'after':[8,9,7,7.9,8,8,7,9,8,4]
}
df=pd.DataFrame(data)
stat,p_value=wilcoxon(df['before'],df['after'])
print(f"Wilcoxon signed-rank statistic:{stat},p-value:{p_value}")
```

kruskal wallis

```python
import pandas as pd
from scipy.stats import kruskal
import numpy as np
d=pd.read_csv('/content/fruits.zip')
d={
'group1':np.random.normal(loc=0,scale=1,size=10),
'group2':np.random.normal(loc=0,scale=1,size=10),
'group3':np.random.normal(loc=0,scale=1,size=10)
}
f_stat,p_value=kruskal(d['group1'],d['group2'],d['group3'])
print(f"kruskal-wallis H statistic:{f_stat},p-value:{p_value}")
```
